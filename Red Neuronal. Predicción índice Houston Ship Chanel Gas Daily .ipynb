```python
filas=len(datos.index)
print(filas)
pred_t1=datos.iloc[filas-1, :]
datos.drop(datos.index[[filas-1]],inplace=True)
pred_t1=pred_t1.drop(['hscgd'])
pred_t1=pred_t1.to_frame().reset_index()
pred_t1=pred_t1.T
pred_t1=pred_t1.to_numpy()
pred_t1=np.delete(pred_t1, 0, axis=0)
print(pred_t1)
```

```python
#Defino el vector de la variable dependiente y la matriz de variables independientes como array
Y=datos['hscgd'].to_numpy()
print(Y)
X=datos.drop(['hscgd'], axis=1).to_numpy()
print(X)
```

```python
#Defino mis conjuntos de entrenamiento y prueba, asigno 50% de las obsrvaciones para test y el resto para entrenemiento
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.5, random_state=12)
print(X_train)
print(X_test)
print(Y_train)
print(Y_test)
```

```python
print(Y_train.shape)
print(X_train.shape)
print(Y_test.shape)
print(X_test.shape)
```

# Ajuste del modelo

```python
#Neuronas en la primer capa
hidden_layer_size=4
output_size = 1

model = tf.keras.Sequential([
    #Primera capa oculta
    tf.keras.layers.Dense(hidden_layer_size,input_dim=24,activation='linear'),
    #Segunda capa oculta
    tf.keras.layers.Dense(output_size ,activation='linear') 
])

model.compile(optimizer = 'Nadam', loss = 'MSE', metrics = ['MSE'])

#entrenamiento
model.fit(X_train, Y_train, epochs = 800,   verbose = 2) 
```

```python
model.weights
```

### Evaluamos y predecimos
En esta sección se insertan los conjuntos de prueba para hacer predicciones y nos arroja el MSE entre las predicciones y las Y observadas, este MSE no debería de ser muy diferente al que se estimó en el ajuste del modelo con los conjuntos de entrenemiento


```python
scores = model.evaluate(X_test, Y_test)
print("\n%s: %.4f%%" % (model.metrics_names[1], scores[1]*100))


print (model.predict(X_test))
```
